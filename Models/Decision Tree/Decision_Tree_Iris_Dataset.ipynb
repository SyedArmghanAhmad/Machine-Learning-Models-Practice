### Decision Tree Classification on the Iris Dataset

#### Code Summary:
This code demonstrates how to use a **Decision Tree Classifier** on the Iris dataset to classify flower species based on their features. The process includes training the model, visualizing the decision tree, and understanding the Gini impurity.

#### Key Steps:
1. **Import Libraries**:
   - `pandas`, `matplotlib.pyplot`: For data manipulation and visualization.
   - `sklearn.datasets.load_iris`: To load the Iris dataset.
   - `sklearn.tree.DecisionTreeClassifier`: To create and train the decision tree model.
   - `sklearn.tree.plot_tree`: To visualize the decision tree.

2. **Load Dataset**:
   - The Iris dataset contains 150 samples divided into three classes: Setosa, Versicolor, and Virginica (50 samples each).

3. **Train the Decision Tree**:
   - The `DecisionTreeClassifier` was trained using the entire dataset.
   - No train-test split was performed here, as this example focuses on visualization.

4. **Visualize the Decision Tree**:
   - Used `tree.plot_tree()` to display the decision tree structure with colored nodes indicating splits and classes.

#### Key Note on Gini Impurity:
- **Gini Impurity Formula**:
  \[
  Gini = 1 - \sum_{i=1}^{n} (p_i)^2
  \]
  \(p_i\): Proportion of samples in class \(i\).

- **Root Node Gini Value**:
  - The Gini value at the root node is **0.667** because the dataset has **three equally represented classes** (\(p_1 = p_2 = p_3 = 0.333\)).
  - Calculation:
    \[
    Gini = 1 - (0.333^2 + 0.333^2 + 0.333^2) = 0.667
    \]
  - This is **normal behavior** for multi-class datasets and not an issue.

- **Behavior**:
  - As the tree splits the data, the Gini impurity decreases, reaching **0** at leaf nodes where all samples belong to one class.

#### Visualization:
- The plotted tree shows how features and thresholds are used at each split.
- Leaf nodes represent the final classification for each sample.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
iris = load_iris()
iris
```

```python
Dtree = DecisionTreeClassifier()
Dtree.fit(iris.data, iris.target)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(Dtree, filled=True)
```
